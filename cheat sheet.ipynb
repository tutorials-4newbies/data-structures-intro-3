```python

class Node:
    def __init__(self, value, next = None):
        self.value = value
        self.next = next
        
    def __str__(self):
        return f"{self.value}"
    
class MyList:
    def __init__(self):
        self.head = None
        self.counter = 0
        
    def is_empty(self)->bool:
        return self.head is None
    
    def size(self):
        return self.counter
    
    def add(self, value):
        node = Node(value)
        tmp_head = self.head
        self.head = node
        self.head.next = tmp_head
        self.counter += 1
    
    def pop(self):
        res = None
        if not self.is_empty():
            
            tmp_head = self.head.next
            res = self.head
            self.head = tmp_head
            self.counter -= 1
        return res
    
    def index(self, value):
        result = self.size()
        temp = self.head
        while temp:
#             print(temp)
            if not temp.value == value:
                result -= result
                temp = temp.next
            else:
                break
            
        return result - 1

l = MyList()
l.add(1)
l.add(2)
print(l.size())
print(l.index(1))
```
